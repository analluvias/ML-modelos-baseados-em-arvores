```python
import pandas as pd

# lendo o csv
df = pd.read_csv("indian_liver_patient_preprocessed.csv")

# mostrando o head
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Liver_disease"])
y = df["Liver_disease"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# instaciar rl
lr = LogisticRegression(random_state=42)

# instanciar knn
knn = KNeighborsClassifier(n_neighbors=27)

# instanciar dt
dt = DecisionTreeClassifier(min_samples_leaf=0.13, random_state=42)

# Define a lista de classificadores
classifiers = [('Logistic Regression', lr), ('K Nearest Neighbours', knn), ('Classification Tree', dt)]
```

vamos avaliar o desempenho dos modelos individualmente primeiro

```python
from sklearn.metrics import accuracy_score

# Iterando pela lista de classificadores
for clf_name, clf in classifiers:

    # akustando no conjunto de treino
    clf.fit(X_train, y_train)

    # predizendo y_pred
    y_pred = clf.predict(X_test)

    # calculando acuracia
    acuracia = accuracy_score(y_test, y_pred)

    # printando avaliação
    print('{:s} : {:.3f}'.format(clf_name, acuracia))
```

Aqui, quem obteve a maior precisão foi o KNN, vamos verificar se ao utilizar um classificador com votação teremos um melhor desempenho

```python
# Importar VotingClassifier
from sklearn.ensemble import VotingClassifier

# instanciar VotingClassifier
vc = VotingClassifier(estimators=classifiers)

# ajustar no conjunto de treino
vc.fit(X_train, y_train)

# avaliar no conjunto de testes
y_pred = vc.predict(X_test)

# calcular acurácia
acuracia = accuracy_score(y_test, y_pred)
print('Voting Classifier: {:.3f}'.format(acuracia))
```

podemos ver que neste caso, o desempenho foi um pouco pior do que o uso do KNN sozinho, alcançando a mesma acurácia da LR
